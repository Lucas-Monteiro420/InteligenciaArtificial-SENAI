```python
# Importando as bibliotecas 
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.dates as mdates
```

```python
# Cria varriável df para o armazenar o DataFrame
df = pd.read_csv('https://raw.githubusercontent.com/danielvieira95/Inteligencia_Artificial_2025/refs/heads/main/bases_de_dados/monitoramento_tempo.csv')
```

```python
# Exibe as quatro primeiras linhas do DataFrame
df.head() 
```

```python
# Exibe as quatro últimas linhas do DataFrame
df.tail()
```

```python
# Importa a biblioteca datetime para fazer a conversão dos dados de data de objeto para datetime
import datetime
df['data'] = pd.to_datetime(df['data'])
```

```python
df.info() # Exibe informações sobre a Base de Dados
```

```python
plt.plot(df['data'], df['temperatura']) # Plota o gráfico de linha da temperatura em função do tempo
```

```python
# Criando código para plotar um gráfico dentro do outro
fig = plt.figure(figsize=(12,5)) # Cria uma figura com tamanho 12x5
eixo = fig.add_axes([0, 0, 1, 1]) # Cria os eixos da figura
eixo.set_ylim(200,500) # Define o limite do eixo y
eixo.plot(df['data'], df['temperatura'], color='blue') # Plota o gráfico de linha da temperatura em função do tempo
eixo.grid(True) # Adiciona a grade ao gráfico

# Definindo o período do gráfico 
eixo.set_xlim(datetime.datetime(2014, 4, 1), datetime.datetime(2014, 6, 1)) # Define o período do gráfico
eixo.set_title('Temperatura em função do tempo', fontsize = 16, pad = 20) # Define o título do gráfico
eixo.set_xlabel('Data', fontsize = 14) # Define o rótulo do eixo x
eixo.set_ylabel('Temperatura', fontsize = 14) # Define o rótulo do eixo y
eixo.legend() # Adiciona a legenda ao gráfico

# Criando o Eixo 2
eixo2 = fig.add_axes([0.35, 0.50, 0.5, 0.3]) # Cria o eixo 2 para o gráfico
# Limites do eixo x
eixo2.xaxis.set_major_locator(mdates.DayLocator(interval = 7)) # Define o número máximo de valores no eixo x
eixo2.xaxis.set_major_formatter(mdates.DateFormatter('%d-%b')) # Define o formato dos valores do eixo x
eixo2.set_xlim(datetime.datetime(2015, 4, 1), datetime.datetime(2015, 5, 1)) # Define o período do gráfico
eixo2.tick_params(axis='x', labelsize = 8) # Rotaciona os valores do eixo x em 45 graus
eixo2.plot(df['data'], df['temperatura'], color='red') # Plota o gráfico de linha da temperatura em função do tempo
eixo2.grid(True) # Adiciona a grade ao gráfico
eixo2.set_title('Temperatura em função do tempo', fontsize = 10, pad = 10) # Define o título do gráfico
eixo2.legend(['Temperatura'], loc = 'best', fontsize = 10) # Adiciona a legenda ao gráfico
```
