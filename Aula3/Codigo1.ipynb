```python
#Importação das bibliotecas
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#Realiza a leitura do arquivo CSV
df = pd.read_csv('basededados/monitoramento_tempo.csv', sep = ',')
```

```python
#Exibe as primeiras linhas do dataframe
df.head() #Exibe as quatro primeiras linhas
```

```python
df.info()
```

```python
import datetime 
df['Data'] = pd.to_datetime(df['Data'])
#Transforma a data de objeto em datetime 
plt.plot(df['Data'], df ['Temperatura'])
plt.figure(figsize = (15,8))
plt.xlabel('Data e Hora')
plt.ylabel('Temperatura')
plt.show()
```

```python

fig = plt.Figure(figsize = (15, 8))  # Cria uma figura
eixo = fig.add_axes([0, 0, 1, 1])  # Adiciona os eixos

eixo.plot(df['Data'], df['Temperatura'] / 10, color='red', lw=0.7)  # Corrigido aqui
eixo.set_title('Temperatura no momento', fontsize=12)
eixo.legend(['Temperatura'], loc='lower right', fontsize=15)
eixo.set_xlabel('Ano', fontsize=14)
eixo.set_ylabel('Ano', fontsize=14)
```

```python
#Selecionando o período da base de dados
fig = plt.figure(figsize=(15,8))
plt.rcParams['font.size'] = '20'
eixo = fig.add_axes([0,0,1,1])
eixo.plot(df['Data'], df['Temperatura'], color= 'b')
eixo.set_xlim(datetime.datetime(2015, 1, 1), datetime.datetime(2015, 12, 31)) #xlim define o valor inicial e o valor final 
eixo.set_title('Temperatura - Ano 2015', fontsize = 25)
eixo.legend(['Temperatura'], fontsize = 20)
eixo.set_xlabel('Ano', fontsize = 20)
```
